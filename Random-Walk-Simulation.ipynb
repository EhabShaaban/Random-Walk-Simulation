```python
# Main Question Out of this?
# What are the odds that you'll reach 60 steps high in a Building?
# Import numpy,matplotlib and set seed
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
```

```python
all_walks=[]

# number of simulations
for i in range(500):
    
    # single random_walk
    random_walk = [0]

    # the for loop where the magic happens
    for x in range(100):
        
        # step is the number of floor
        step = random_walk[-1]
        
        # Roll the dice
        dice = np.random.randint(1,7)

        # Determine next step
        if dice <= 2:
            #never goes below zero
            step = max(0, step - 1)
            
        elif dice <= 5:
            step = step + 1
        # if your dice gave you six (its your lucky day BUD!, kind off..)    
        else:
            step = step + np.random.randint(1,7)
            
        # install clumsiness with 0.1%
        if np.random.rand() <= 0.001 :
            step = 0
            
# The next two lines of code is that where we can put the walks into a list object
# that we can at the end look at the every walk end and see when the simulation ended.

        # appending step to random_walk
        random_walk.append(step)
        
    # Append random_walk to all_walks
    all_walks.append(random_walk)

np_aw_t = np.transpose(np.array(all_walks))
np_aw = np.array(all_walks)
np_aw_t = np.transpose(np_aw)

# Selecting the end of each walk
ends = np_aw_t[-1]

# Plotting Data
plt.hist(ends)
plt.title("End of All Random Walks")
plt.tight_layout()

plt.show()

plt.plot(np_aw)
plt.title("All Random Walks")
plt.tight_layout()
plt.show()

plt.plot(np_aw_t)
plt.title("Transposed Random Walks")
plt.tight_layout()
plt.show()

plt.plot(random_walk)
plt.title("Random Walk")
plt.tight_layout()
plt.show()

#you can count the number of integers in ends that are
#greater than or equal to 60 and divide that number by 500, the total number of simulations.
#then you'll answer the question.. YAYYYY.
```
